#PIPELINE NLP: Formalizzazione di Informazioni da Testo

##Implementazione di una Pipeline NLP in Python

##Carichiamo gli strumenti per l'analisi del testo: Libreria NLTK

```python
import nltk
nltk.download('all')
```

##Livello 1 : Token
Abbiamo il Livello 0
il testo

vogliamo arrivare al livello 1 della pipeline: estraiamo le parole
Eseguamo lo split del testo in parole

```python
#scarico il pacchetto con la punteggiatura, che sarà  effeettauto per eseguire la tokenizzazione
nltk.download('punkt')
#Livello 1 : estraggo le parole
token=nltk.word_tokenize(Testo_pagina1)
```

```python
token
```

##Livello 2: Frasi
Abbiamo il Livello 1
il testo

vogliamo arrivare al livello 2 della pipeline: estraiamo le frasi

```python
frasi=nltk.sent_tokenize(Testo_pagina1)
```

```python
frasi
```

##Livello 3: Parte del Discorso POS - Categoria Grammaticale
In questo livello si associa, con un certo tasso di errore, la categoria grammaticale ai token estratti dal testo

```python
POS_tag=nltk.pos_tag(token)
```

Le categorie grammaticali (POS) sono rappresentate mediante opportuni acronimi come riportato nella tabella che segue
